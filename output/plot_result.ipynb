## Plot

```python
colors = ['b', 'g', 'r', 'c', 'm', 'y']
```

### Accuracy on testing dataset

```python

import matplotlib.pyplot as plt
import numpy as np

categories = ("MNIST-2", "FMNIST-2", "FMNIST-3", "Syn-4")
accuracies = {
    'Naive Bayes': np.round(naive_bayes_test_acc, 2),
    'SVM': np.round(svm_test_acc, 2),
    'SimpleANN': np.round(simpleANN_test_acc, 2),
    'ComplexANN': np.round(complexANN_test_acc, 2),
    'SimpleVQC': np.round(simpleVQC_test_acc, 2),
    'ComplexVQC': np.round(complexVQC_test_acc, 2),
}

x = np.arange(len(categories))  # the label locations
width = 0.1  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained', figsize=(12, 8))

for attribute, measurement in accuracies.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute, color=colors[multiplier])
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Accuracy')
# ax.set_title('Testing accuracy')
ax.set_xlabel("Dataset")
ax.set_xticks(x + width+0.2, categories)
ax.legend(loc='best', ncols=6)
ax.set_ybound(upper=1.2)
fig.savefig(os.path.join(DATA_DIR, 'testing_accuracy.pdf'), bbox_inches='tight')
fig.savefig(os.path.join(DATA_DIR, 'testing_accuracy.png'), bbox_inches='tight')
plt.show()
```
